```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Data_set 2 - Copy.csv")
```

```python
#Task 3: Descriptive Statistics

print(df.select_dtypes(include='number').columns.tolist())
```

```python
numerical_df = df.select_dtypes(include='number')

print("Mean:")
print(numerical_df.mean().round(2))

print("\nMedian:")
print(numerical_df.median().round(2))

print("\nStandard Deviation:")
print(numerical_df.std().round(2))
```

```python
print(numerical_df.describe().round(2))
```

```python
plt.figure(figsize=(10, 5))
means = numerical_df.drop('age', axis=1).mean().round(2)
plt.bar(means.index, means.values, color='steelblue')
plt.title("Mean Preference Ranking per Investment Type (1=Most Preferred)")
plt.xlabel("Investment Type")
plt.ylabel("Mean Rank")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("task3_descriptive_stats.png")
plt.show()
```

```python
#Task 4: Most Preferred Investment Avenue

avenue_counts = df['Avenue'].value_counts()
print(avenue_counts)
print("\nPercentages:")
print(df['Avenue'].value_counts(normalize=True).mul(100).round(1))
```

```python
plt.figure(figsize=(8, 5))
plt.bar(avenue_counts.index, avenue_counts.values, color=['steelblue', 'salmon', 'mediumseagreen', 'mediumpurple'])
plt.title("Most Preferred Investment Avenue")
plt.xlabel("Avenue")
plt.ylabel("Number of Respondents")
plt.tight_layout()
plt.savefig("task4_investment_avenue.png")
plt.show()
```
