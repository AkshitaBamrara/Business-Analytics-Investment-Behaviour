```python
#Task 1: Data Overview

import pandas as pd

df = pd.read_csv("Data_set 2 - Copy.csv")

print("Shape:", df.shape)

print("\nDataset Info:")
print(df.info())

print("\nFirst 5 rows:")
print(df.head())

print("\nMissing values:")
print(df.isnull().sum())
```

```python
#Task 2: Gender Distribution

import matplotlib.pyplot as plt

gender_counts = df['gender'].value_counts()
print(gender_counts)

plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['steelblue', 'hotpink'])
plt.title("Gender Distribution")
plt.show()

plt.figure(figsize=(6, 4))
plt.bar(gender_counts.index, gender_counts.values, color=['steelblue', 'hotpink'])
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.savefig("task2_gender_distribution.png")
plt.show()
```
