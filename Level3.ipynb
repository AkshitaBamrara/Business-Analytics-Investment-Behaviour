```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Data_set 2 - Copy.csv")
```

```python
#Task 5: Reasons for Investment

reason_columns = ['Reason_Equity', 'Reason_Mutual', 'Reason_Bonds', 'Reason_FD']

for col in reason_columns:
    print(f"\n{col}:")
    print(df[col].value_counts())
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle("Reasons for Investment by Type", fontsize=14)

for ax, col in zip(axes.flatten(), reason_columns):
    counts = df[col].value_counts()
    ax.bar(counts.index, counts.values, color='steelblue')
    ax.set_title(col)
    ax.set_ylabel("Count")
    ax.tick_params(axis='x', rotation=15)

plt.tight_layout()
plt.savefig("task5_reasons.png")
plt.show()
```

```python
#Task 6: Savings Objectives

print("Savings Objectives:")
print(df['What are your savings objectives?'].value_counts())

print("\nInvestment Purpose:")
print(df['Purpose'].value_counts())
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
fig.suptitle("Savings Objectives & Investment Purpose", fontsize=14)

obj_counts = df['What are your savings objectives?'].value_counts()
axes[0].bar(obj_counts.index, obj_counts.values, color='mediumpurple')
axes[0].set_title("Savings Objectives")
axes[0].set_ylabel("Count")
axes[0].tick_params(axis='x', rotation=15)

pur_counts = df['Purpose'].value_counts()
axes[1].bar(pur_counts.index, pur_counts.values, color='mediumseagreen')
axes[1].set_title("Investment Purpose")
axes[1].set_ylabel("Count")
axes[1].tick_params(axis='x', rotation=15)

plt.tight_layout()
plt.savefig("task6_savings_objectives.png")
plt.show()
```
